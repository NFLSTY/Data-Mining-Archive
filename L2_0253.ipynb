6. a. Download dataset Video Games dari dataset amazon (https://amazon-reviews-2023.github.io/). Download yang bagian review saja <br>
b. Hitung jumlah user<br>
c. hitung jumlah item<br>
d. tampilkan jumlah rating<br>
e. tampilkan user yang paling banyak memberi rating<br>
f. tampilkan item yang paling banyak di rating<br>
g. tampilkan user yang paling sedikit meberi rating<br>


```python
import pandas as pd
import json
import gzip

# 6a. Download dataset
file_path = "/content/drive/MyDrive/Semester 5/Data Mining/Week2/Video_Games.jsonl.gz"
# Karena tiap baris adalah JSON, kita baca manual
data = []
with gzip.open(file_path, "rt", encoding="utf-8") as f:
    for line in f:
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError:
            continue  # skip baris rusak jika ada

df = pd.DataFrame(data)
df
```

```python
print("Kolom yang tersedia:", df.columns.tolist())
```

```python
# 6b. Hitung jumlah user
num_users = df["user_id"].nunique()
print("Jumlah user unik:", num_users)
```

```python
# 6c. Hitung jumlah item
num_items = df["asin"].nunique()
print("Jumlah item unik:", num_items)
```

```python
# 6d. Tampilkan jumlah rating
num_ratings = df["rating"].count()
print("Jumlah rating:", num_ratings)
```

```python
# 6e. Tampilkan user yang paling banyak memberi rating
user_counts = df["user_id"].value_counts()
most_active_user = user_counts.idxmax()
most_active_user_count = user_counts.max()
print("User yang paling banyak memberi rating:", most_active_user)
print("Jumlah rating yang diberikan:", most_active_user_count)
```

```python
# 6f. Tampilkan item yang paling banyak di rating
item_counts = df["asin"].value_counts()
most_rated_item = item_counts.idxmax()
most_rated_item_count = item_counts.max()
print("Item yang paling banyak di-rating:", most_rated_item)
print("Jumlah rating yang diterima:", most_rated_item_count)
```

```python
# 6g. Tampilkan user yang paling sedikit meberi rating
least_active_user = user_counts.idxmin()
least_active_user_count = user_counts.min()
print("User yang paling sedikit memberi rating:", least_active_user)
print("Jumlah rating yang diberikan:", least_active_user_count)
```
